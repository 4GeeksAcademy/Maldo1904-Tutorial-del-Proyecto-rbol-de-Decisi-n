```python
cols_a_corregir = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df2 = df[(df[cols_a_corregir] != 0).all(axis=1)]
```

```python
stats_Insuline = df2['Insulin'].describe() 
stats_Insuline
```

```python
import pandas as pd
import numpy as np


Insuline_iqr = stats_Insuline['75%'] - stats_Insuline['25%']
limite_superior_Insuline = stats_Insuline['75%'] + 1.5 * Insuline_iqr
limite_inferior_Insuline = stats_Insuline['25%'] - 1.5 * Insuline_iqr

# Ver los lÃ­mites calculados
print(f'El lÃ­mite superior es {round(limite_superior_Insuline, 2)}')
print(f'El lÃ­mite inferior es {round(limite_inferior_Insuline, 2)}')

df2 = df2[(df2['Insulin'] >= limite_inferior_Insuline) & (df2['Insulin'] <= limite_superior_Insuline)]


print(df2.head())  
```

```python
stats_BMI = df2['BMI'].describe() 
stats_BMI
```

```python
import pandas as pd
import numpy as np

BMI_iqr = stats_BMI['75%'] - stats_BMI['25%']
limite_superior_BMI = stats_BMI['75%'] + 1.5 * BMI_iqr
limite_inferior_BMI = stats_BMI['25%'] - 1.5 * BMI_iqr

# Ver los lÃ­mites calculados
print(f'El lÃ­mite superior es {round(limite_superior_BMI, 2)}')
print(f'El lÃ­mite inferior es {round(limite_inferior_BMI, 2)}')

df2 = df2[(df2['BMI'] >= limite_inferior_BMI) & (df2['BMI'] <= limite_superior_BMI)]


```

```python
stats_DiabetesPedigreeFunction= df2['DiabetesPedigreeFunction'].describe() 
stats_DiabetesPedigreeFunction
```

```python
import pandas as pd
import numpy as np

DiabetesPedigreeFunction_iqr = stats_DiabetesPedigreeFunction['75%'] - stats_DiabetesPedigreeFunction['25%']
limite_superior_DiabetesPedigreeFunction = stats_DiabetesPedigreeFunction['75%'] + 1.5 * DiabetesPedigreeFunction_iqr
limite_inferior_DiabetesPedigreeFunction = stats_DiabetesPedigreeFunction['25%'] - 1.5 * DiabetesPedigreeFunction_iqr

# Ver los lÃ­mites calculados
print(f'El lÃ­mite superior es {round(limite_superior_DiabetesPedigreeFunction, 2)}')
print(f'El lÃ­mite inferior es {round(limite_inferior_DiabetesPedigreeFunction, 2)}')

df2 = df2[(df2['DiabetesPedigreeFunction'] >= limite_inferior_DiabetesPedigreeFunction) & (df2['DiabetesPedigreeFunction'] <= limite_superior_DiabetesPedigreeFunction)]


```

```python
from sklearn.model_selection import train_test_split

X = df2.drop('Outcome', axis=1) 
y = df2['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=42)
model.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Suponiendo que 'model' es un Ã¡rbol de decisiÃ³n previamente entrenado
fig = plt.figure(figsize=(15, 15))

tree.plot_tree(model, feature_names=list(X_train.columns), class_names=["0", "1"], filled=True)

plt.show()
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
model1=DecisionTreeClassifier(random_state=42,criterion='entropy')
model2=DecisionTreeClassifier(random_state=42,criterion='log_loss')

model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

# Suponiendo que 'model' es un Ã¡rbol de decisiÃ³n previamente entrenado
fig = plt.figure(figsize=(15, 15))

tree.plot_tree(model1, feature_names=list(X_train.columns), class_names=["0", "1"], filled=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

fig = plt.figure(figsize=(15, 15))

tree.plot_tree(model2, feature_names=list(X_train.columns), class_names=["0", "1"], filled=True)

plt.show()
```

```python
y_pred_entropy=model1.predict(X_test)
print(accuracy_score(y_test,y_pred_entropy))

y_pred_log_loss=model2.predict(X_test)
print(accuracy_score(y_test,y_pred_log_loss))
```
